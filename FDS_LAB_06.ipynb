```python
# Making DataFrame :
import numpy as np
import pandas as pd
```

```python
num = np.array([[85,78,92],[90,82,94],[88,80,90],[87,79,93]])
A = pd.DataFrame(num,columns=['M1','M2','M3'])
A
```

```python
# The NULL Hypothesis : Ho => Mean(M1) = Mean(M2) = Mean(M3)
# The Alternate Hypothesis : Ha => Mean(M1) != Mean(M1) != Mean(M1)
```

```python
# One - Way ANOVA :
# Step 1 + 2 : Mean of all columms and Overall Mean -
def mean_cal(x):
    return sum(x)/len(x)

# Step 3 : Calculating SSB -
def SSB_cal(av,om,r):
    sum = 0
    for i in av:
      sum += (i-om)**2
    return r*sum

# Step 4 : Calculating SST -
def SST_cal(arr,om):
    sum = 0
    for i in arr:
      sum += (i-om)**2
    return sum

# Step 5 : Calculating SSW -
def SSW_cal(ssb_val,sst_val):
  return sst_val-ssb_val

# Step 6 : Calculating MSB -
def MSB_cal(ssb_val,df1):
  return ssb_val/df1

# Step 7 : Calculating MSW -
def MSW_cal(ssw_val,df2):
  return ssw_val/df2

# Step 8 : Calculating Fstatic -
def Fstatic_cal(msb_val,msw_val):
  return msb_val/msw_val
```

```python
# Step 9 : F-Critical value calculation at alpha = 0.05 at df1 and df2
import scipy.stats
def f_table_cal(alpha,df1,df2):
  return scipy.stats.f.ppf(q=1-alpha, dfn=df1, dfd=df2)
```

```python
num = np.array([[85,78,92],[90,82,94],[88,80,90],[87,79,93]])
A = pd.DataFrame(num,columns=['M1','M2','M3'])
A

# Example usage (assuming you have already calculated df1 and df2)
alpha = 0.05
df1 = 2  # Example value, replace with your actual df1
df2 = 9  # Example value, replace with your actual df2
f_critical = f_table_cal(alpha, df1, df2)
print(f"F-critical value: {f_critical}")
```
